# Requirement R1

```python
import pandas as pd
sk = pd.read_csv('sk.csv')
sk
```

```python
sk.isnull().sum()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

numeric_features = sk.select_dtypes(include=['float64', 'int64']).columns.tolist()
categorical_features = sk.select_dtypes(include=['object']).columns.tolist()

transformers = [('one_hot_encoder', OneHotEncoder(), categorical_features)]
ct = ColumnTransformer(transformers, remainder='passthrough')

# Fit and transforming
new_sk = ct.fit_transform(sk)
```

```python
x = new_sk[:, :-1]
y = new_sk[:, -1]
```

```python
# minmaxscaled x_train
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
x = mm.fit_transform(x)
```

Splitting data into 60% for training , 20% for testing , 20% for validation

```python
from sklearn.model_selection import train_test_split

x_train, x_temp, y_train, y_temp = train_test_split(x, y, test_size=0.4, random_state=0)
x_test, x_val, y_test, y_val = train_test_split(x_temp, y_temp, test_size=0.5, random_state=0)
```

```python
from collections import Counter
Counter(y_train)
```

# Requirement R2

```python
from sklearn.linear_model import Perceptron
ppt1 = Perceptron()
ppt1.fit(x_train,y_train)
```

```python
y_pred = ppt1.predict(x_test)
```

```python
from sklearn.metrics import f1_score
f1_class_1 = f1_score(y_test, y_pred, pos_label=1)
f1_class_2 = f1_score(y_test, y_pred, pos_label=2)

# Printed F1 scores for each class
print(f" Class 1: {f1_class_1}")
print(f" Class 2: {f1_class_2}")
```

# Requirement R7

```python
categorical_mask = new_df.dtypes == object
```

```python
from sklearn.model_selection import GridSearchCV
from imblearn.over_sampling import SMOTENC
from imblearn.pipeline import Pipeline
from sklearn.linear_model import Perceptron

# Define a pipeline
model = Perceptron()
resampling = SMOTENC(categorical_features = categorical_mask)
pipeline = Pipeline([('SMOTENC', resampling), ('Perceptron', model)])

# Define parameter grid
param_grid = {
    'SMOTENC__sampling_strategy': [0.5, 0.75, 1.0],
    'SMOTENC__k_neighbors': [3, 5, 7]
}

# Perform grid search
grid_search = GridSearchCV(pipeline, param_grid, scoring='f1', cv=5)
grid_search.fit(a_val, b_val)

```

```python
print("The optimal combination of oversampling ratio and n values is:", grid_search.best_params_)

```

```python
# trained the  model with the best parameters
best_params = grid_search.best_params_
resampling = SMOTENC(categorical_features=categorical_mask, sampling_strategy=best_params['SMOTENC__sampling_strategy'], k_neighbors=best_params['SMOTENC__k_neighbors'])
a_resampled, b_resampled = resampling.fit_resample(a_train, b_train)
model = Perceptron()
model.fit(a_resampled, b_resampled)


b_pred = model.predict(a_test)
f1_class_1 = f1_score(b_test, b_pred, pos_label=1)
f1_class_2 = f1_score(b_test, b_pred, pos_label=2)

print(f"Class 1: {f1_class_1}")
print(f"Class 2: {f1_class_2}")
```
